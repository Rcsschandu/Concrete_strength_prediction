# REMOVING OUTLIERS FROM THE DATA.

```python
def out_rem(column):
    q3 = c[column].quantile(q=0.75)
    q1 = c[column].quantile(q=0.25)
    iqr = q3 - q1
    upper_bound = q3 + 1.5*iqr
    lower_bound = q1 - 1.5*iqr
    out_idx = c[(c[column]>upper_bound) | (c[column]<lower_bound)].index
    print(c[column].name)
    print(q3,"-",q1,"=",iqr)
    print(upper_bound)
    print(lower_bound)
    print(out_idx)
    c.drop(labels=(out_idx),axis=0,inplace=True)
```

# Training Using STATS MODEL

```python
y_train = df_train.pop('Concrete_compressive_strength')
X_train = df_train
```

```python
X_train.head()
```

```python
import statsmodels.api as sm

# Add a constant to get an intercept
X_train_sm = sm.add_constant(X_train)

# Fit the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()
```

## If p-value of any feature is less than 0.05, that means feature is statistically significant

# Variance Infllation Factor (VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif
```

```python
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
